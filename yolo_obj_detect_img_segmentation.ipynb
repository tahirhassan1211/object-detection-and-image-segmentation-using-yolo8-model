```python
from ultralytics import YOLO
import gradio as gr
import cv2
import numpy as np
from PIL import Image
```

```python
det_model = YOLO("yolov8n.pt")        # Object Detection
seg_model = YOLO("yolov8n-seg.pt")    # Image Segmentation
```

```python
def yolo_inference(image, task):
    if image is None:
        return None

    # Convert PIL → OpenCV
    image = np.array(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    if task == "Object Detection":
        results = det_model(image)
    else:
        results = seg_model(image)

    # Plot results
    annotated_image = results[0].plot()

    # Convert back to PIL
    annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
    return Image.fromarray(annotated_image)
```

```python
interface = gr.Interface(
    fn=yolo_inference,
    inputs=[
        gr.Image(type="pil", label="Upload Image"),
        gr.Radio(
            ["Object Detection", "Image Segmentation"],
            label="Select Task",
            value="Object Detection"
        )
    ],
    outputs=gr.Image(label="Result"),
    title="YOLOv8 Object Detection & Image Segmentation",
    description="Upload an image and choose between Object Detection or Image Segmentation using YOLOv8."
)
```

```python
interface.launch(debug=True)
```
